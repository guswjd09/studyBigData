##### 회귀 분석을 위한 statsmodle 패키지 import

```python
from statsmodels.formula.api import ols, glm
```

```python
# type을 지정해서 두 그룹을 따로 뽑아서 분리한 Series를 만듬
# loc는 location의 줄임말이고 콤마 앞에 있는 것이 row, 뒤에 있는 것이 columns

red_wine_quality = wine.loc[wine['type'] == 'red', 'quality']
white_wine_quality = wine.loc[wine.type == 'white', 'quality']
```

```python
red_wine_quality
```

```python
white_wine_quality
```

```python
# t-검정 : 두 그룹 사이의 차이
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var = False)
```

```python
# 선형회귀분석 linear regression
Rformula = 'quality ~ '
for c in list(wine.columns)[1:-2]:
    Rformula += c + ' + '
Rformula += list(wine.columns)[-2]

Rformula
```

```python
# 선형회귀분석 linear regression
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
regre_result = ols(Rformula, data=wine).fit()
regre_result.summary()
```

```python
regre_result = glm(Rformula, data=wine).fit()
regre_result.summary()
```

#### OLS모델로 계산된 회귀분석 결과로 새 샘플 와인에 대한 품질등급 예측

```python
# quality, type 제거, 컬럼 알파벳순으로 정렬
sample1 = wine[wine.columns.difference(['quality', 'type'])]
sample1
```

```python
# red_wine에서만 예측 
sample1 = sample1[0:5][:]
sample1
```

```python
# 예측된 값
sample1_predict = regre_result.predict(sample1)
sample1_predict
```

```python
# 원래 데이터의 0번부터 5번까지 quality만 보겠다
# 기존의 데이터가 가진 값과 위에서 sample1_predict(독립변수만 분리한 1~5행의 값)으로 실제로 회귀식이 얼마나 잘 예측하는지 비교해볼 수 있다.
wine[0:5]['quality']
```

```python
# 새로운 와인에서 추출된 데이터를 가지고 등급 예측
data = {'fixed_acidity':[8.5, 8.1], 
        'volatile_acidity':[0.8, 0.5],
        'citric_acid':[0.3,0.4],
        'residual_sugar':[6.1, 5.8],
        'chlorides':[0.055, 0.04],
        'free_sulfur_dioxide':[30.0, 31.0],
        'total_sulfur_dioxide':[98.0, 99],
        'density':[0.996, 0.91],
        'pH':[3.25, 3.01],
        'sulphates':[0.4, 0.35],
        'alcohol':[9.0, 0.88]}

new_sample = pd.DataFrame(data, columns=sample1.columns)
new_sample
```

```python
new_sample_predict = regre_result.predict(new_sample)
new_sample_predict
```
